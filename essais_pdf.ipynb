```python
from typing import Any, Iterable
from bs4.element import Tag
import textdistance
import nltk
import unidecode
from copy import copy
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
# from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from xvfbwrapper import Xvfb
import rispy
import fitz
import string as str_cst
import re

class Document():
    def __new__(cls, document_extract:Tag, driver, database:str) :
        document_retrieve = DocumentRetriever(document_extract, driver)
        if not document_retrieve.pdf_file:
            return None

        else:
            instance = super().__new__(cls)
            instance.database = database
            instance.__document_retrieve = document_retrieve
            return instance
        
        # self = cls
        
        # self.doc_type = "pdf"

        # self.database = database

        # self.pdf        = document_retrieve.pdf_file
        # self.es_pdf_id  = None
        # self.type       = document_retrieve.biblio_data.get("type_of_reference")
        # self.title      = document_retrieve.biblio_data.get("primary_title")
        # self.authors    = document_retrieve.biblio_data.get('first_authors')
        # self.vol        = document_retrieve.biblio_data.get('volume')
        # self.j_name     = document_retrieve.biblio_data.get('journal_name')
        # self.nb         = document_retrieve.biblio_data.get('number')
        # self.pages      = "-".join([document_retrieve.biblio_data.get('start_page', ""), document_retrieve.biblio_data.get('end_page', "")])
        # self.issn       = document_retrieve.biblio_data.get('issn')
        # self.pub_year   = document_retrieve.biblio_data.get('publication_year')
        # self.publisher  = document_retrieve.biblio_data.get('publisher')

        # parsed_pdf = PDFParser(pdf_file=document_retrieve.pdf_file)

        # self.paragraphs = parsed_pdf.paragraphs

        # return self

        

    def __init__(self, *args, **kwargs) -> None:
        # document_retrieve = DocumentRetriever(document_extract, driver)
        # if not document_retrieve.pdf_file:
        #     return None
        
        self.doc_type = "pdf"

        # self.database = database

        self.pdf        = self.__document_retrieve.pdf_file
        self.es_pdf_id  = None
        self.type       = self.__document_retrieve.biblio_data.get("type_of_reference")
        self.title      = self.__document_retrieve.biblio_data.get("primary_title")
        self.authors    = self.__document_retrieve.biblio_data.get('first_authors')
        self.vol        = self.__document_retrieve.biblio_data.get('volume')
        self.j_name     = self.__document_retrieve.biblio_data.get('journal_name')
        self.nb         = self.__document_retrieve.biblio_data.get('number')
        self.pages      = "-".join([self.__document_retrieve.biblio_data.get('start_page', ""), self.__document_retrieve.biblio_data.get('end_page', "")])
        self.issn       = self.__document_retrieve.biblio_data.get('issn')
        self.pub_year   = self.__document_retrieve.biblio_data.get('publication_year')
        self.publisher  = self.__document_retrieve.biblio_data.get('publisher')

        parsed_pdf = PDFParser(pdf_file=self.__document_retrieve.pdf_file)

        self.paragraphs = parsed_pdf.paragraphs

        del self.__document_retrieve

    def __get_document_paragraphs(self):
        infos = {
            "database"          :   self.database,
            "es_pdf_id"         :   self.es_pdf_id,
            "type"              :   self.type,
            "title"             :   self.title,
            "authors"           :   self.authors,
            "volume"            :   self.vol,
            "journal_name"      :   self.j_name,
            "number"            :   self.nb,
            "pages"             :   self.pages,
            'issn'              :   self.issn,
            "publication_year"  :   self.pub_year,
            "publisher"         :   self.publisher,
        }

        document = dict()

        for info in infos:
            if infos[info]:
                document.update({
                    info    :   infos[info]
                })
        
        for paragraph in self.paragraphs :
            document_out = copy(document)
            document_out.update(paragraph)
            yield document_out
    
    def get_pdf(self) -> dict | None :
        if self.pdf:
            infos = {
                "database"          :   self.database,
                "type"              :   self.type,
                "title"             :   self.title,
                "authors"           :   self.authors,
                "volume"            :   self.vol,
                "journal_name"      :   self.j_name,
                "number"            :   self.nb,
                "pages"             :   self.pages,
                'issn'              :   self.issn,
                "publication_year"  :   self.pub_year,
                "publisher"         :   self.publisher,
                'pdf_file'          :   self.pdf,
            }
            document = dict()

            for info in infos:
                if infos[info]:
                    document.update({
                        info    :   infos[info]
                    })
            
            return document
        else :
            return None
    
    # def get_paragraphs_to_list(self):
        
    #     return list(self.__get_document_paragraphs())

    def __call__(self, *args: Any, **kwds: Any) -> Any:

        return self.__get_document_paragraphs()


class DocumentRetriever():
    def __init__(self, document_extract:Tag, driver) -> None:

        self.driver = driver

        self.result_id = document_extract.attrs['data-cid']

        self.__get_pdf_file_if_exist(document_extract = document_extract)

        if self.pdf_file :
            self.biblio_data = self.__get_bilio_info()


    def __get_bilio_info(self) -> dict:

        self.driver.find_element(by=By.XPATH, value=f"//div[@data-cid='{self.result_id}']//a[@aria-controls='gs_cit']").click()
        WebDriverWait(self.driver, 5.0).until(lambda d: self.driver.find_element(by=By.XPATH, value="//a[@class='gs_citi' and contains(., 'RefMan')]"))
        
        ris_citation = self.driver.find_element(by=By.XPATH, value="//a[@class='gs_citi' and contains(., 'RefMan')]")
        # page_parsing = BeautifulSoup(essai.page_source, 'html.parser')
        # # pprint(self.driver.page_source)
        # # pprint(page_parsing)
        # ref_page = page_parsing.find("div" , attrs={'id' : "gs_cit-bdy"})
        # ref_page = self.driver.find_element(by=By.ID, value="gs_cit").get_attribute('innerHTML')
        # print(ref_page)
        # ref_page = BeautifulSoup(ref_page, 'html.parser')
        ris_citation = BeautifulSoup(ris_citation.get_attribute('outerHTML'), 'html.parser')
        ris_file_url = ris_citation.find("a", string="RefMan").get('href')

        try:
            resp = requests.get(url=ris_file_url)
        except:
            e_text = f"Erreur de récupération des références biblio"
            raise ConnectionError()
        if 200 >= resp.status_code < 300:
            biblio = rispy.loads(resp.content.decode(encoding='utf-8'))
            self.driver.find_element(by=By.XPATH, value=f"//div[@class='gs_md_d gs_md_ds gs_ttzi gs_vis']//a[@id='gs_cit-x']").click()
        else :
            e_text = f"Erreur lors de la récupération des données bibliographiques: code {resp.status_code}, reason {resp.reason}"
            raise RuntimeError(e_text)
        
        if isinstance(biblio, list):
            biblio = biblio[0]
            
        return biblio
    
    def __download_pdf_file(self, url):
        try :
            resp = requests.get(url=url)
        except Exception as e:
            e_text = f"Erreur lors du téléchargement du fichier pdf"
            raise RuntimeError(e_text) from e
        
        if 200 >= resp.status_code < 300:
            return resp.content
        else:
            e_text = f"Erreur lors du téléchargement du fichier pdf : code {resp.status_code}, reason {resp.reason}"
            raise RuntimeError(e_text)
    
    def __get_pdf_file_if_exist(self, document_extract:Tag):
        
        pdf_url = document_extract.find("a", attrs={'data-clk-atid' : self.result_id})
        if pdf_url :
            pdf_url = pdf_url.get('href')
            self.pdf_file = self.__download_pdf_file(url=pdf_url)
        else :
            self.pdf_file = None
            
# class QueryNERExtractor():
#     __nlp = spacy.load("en_core_sci_lg")

#     def __init__(self, query:str) -> None:

#         if not isinstance(query, str):
#             e_text = f"Le paramètre query doit être une stirng."
#             raise ValueError(e_text)
        
#         self.analyzed_query = self.__nlp(text=query)
    
#     def __call__(self, *args: Any, **kwds: Any) -> Any:
#         return self.analyzed_query.ents


class ScholarHandler():
    
    def __init__(self) -> None:
        self.scholar_base_url = "https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5"
        
        # self.vdisplay = Xvfb()
        # self.vdisplay.start()

        # chrome_options = webdriver.ChromeOptions()
        # chrome_options.add_argument("--headless=new")

        # self.driver = webdriver.Chrome(options=chrome_options)
        # "q=vector+base&btnG="
    
    # def __doc_extractor(self, html_page):
    #     page = BeautifulSoup(html_page, 'html.parser')
    #     papers = soup.find_all('div', attrs={'class' : "gs_ri"})
    
    def __enter__(self):
        # self.__vdisplay = Xvfb()
        # self.__vdisplay.start()

        chrome_options = webdriver.ChromeOptions()
        chrome_options.add_argument("--headless=new")

        self.driver = webdriver.Chrome(
            options=chrome_options
            )
        return self
    
    def __exit__(self):
        self.driver.close()
        # self.__vdisplay.stop()

    def __place_on_page(self, url:str):

        if not isinstance(url, str):
            e_text = "L'URL de la page doit être une string."
            raise ValueError(e_text)

        self.driver.get(url = url)
        self.results_page = BeautifulSoup(self.driver.page_source, 'html.parser')

    def __get_research_pages_url(self, nb_pages_ret):

        navigator = self.results_page.find("div", attrs={'id' : "gs_n", 'role' : "navigation"})
        pages = navigator.find_all("a", limit=nb_pages_ret)

        return [page.get("href") for page in pages]
    
    def __get_page_results(self):

        return self.results_page.find_all('div', attrs={'class' : "gs_r gs_or gs_scl"})
    
    def __get_documents_from_current_page(self):

        # page = BeautifulSoup(self.driver.page_source, 'html.parser')
        
        results = self.__get_page_results()

        documents = list()

        if results :
            for result in results :
                documents.append(Document(document_extract=result, driver=self.driver, database="google scholar"))
            
            return [document for document in documents if document]


    def __get_documents_from_research(self, query:list, nb_pages_ret):
        # self.driver = webdriver.Chrome()
        # self.driver.get(url = f"{self.scholar_base_url}&q={'+'.join(query)}&btnG=''")
        # self.results_page = BeautifulSoup(driver.page_source, 'html.parser')

        self.search_url = f"{self.scholar_base_url}&q={'+'.join(query)}&btnG=''"
        self.__place_on_page(url=self.search_url)

        # navigator = result_page.find("div", attrs={'id' : "gs_n", 'role' : "navigation"})
        # pages = navigator.find_all("a", limit=2)
        if nb_pages_ret :
            self.__pages = self.__get_research_pages_url(nb_pages_ret = nb_pages_ret)
        else :
            self.__pages = None
        

        documents = list()
        documents.extend(self.__get_documents_from_current_page())

        if self.__pages :
            for result in self.__pages:
                self.__place_on_page(result)
                documents.extend(self.__get_documents_from_current_page())
        
        self.parsed_pdf_documents = documents
    
    def search(self, query_terms:Iterable, nb_pages:int):

        # if not isinstance(query, str) :
        #     e_text = f"Le paramètre query doit être une chaîne de caractère."
        #     raise ValueError(e_text)
        
        if not isinstance(query_terms, Iterable) and not isinstance(query_terms, str) :
            e_text = "Le paramètre query_terms doit être un iterable (liste, tuple, set ...) et non une chaine de caractère"
            raise ValueError(e_text)

        if not isinstance(nb_pages, int) and nb_pages > 0 :
            e_text = f"Le paramètre nb_pages doit être un entier suppérieur à 0 : valeur entrée {str(nb_pages)}"
            raise ValueError(e_text)

        self.__get_documents_from_research(query=query_terms, nb_pages_ret=nb_pages)


class PDFParser():
    __tversky = textdistance.Tversky(ks=(1, 0.5))
    __stopwords = set(nltk.corpus.stopwords.words('english'))
    __english_words = set(nltk.corpus.words.words())
    def __init__(self, pdf_file) -> None:

        self.__file = pdf_file

        self.__toc_titles = None
        self.__toc_titles_hierarchy = None

        self.__parsed_pdf = fitz.Document(stream=self.__file)

        self.__parse_toc()

        self.__parse_pdf()

    def __parse_toc(self):
        toc = self.__parsed_pdf.get_toc()
        
        if toc :
            self.__toc_titles = set()
            self.__toc_titles_hierarchy = dict()
            for title in toc:
                self.__toc_titles.add(title[1].strip().lower())
                self.__toc_titles_hierarchy.update({
                        tuple(self.__preprocessing_text(title[1])):   {
                        'title'     :  title[1].strip().lower(),
                        'hierarchy' :   title[0]}
                })
    
    def __preprocessing_text(self, string:str) -> list :
        out_text = unidecode.unidecode(string)
        out_text = ''.join([c for c in out_text if c not in str_cst.punctuation])
        out_text = ''.join([i for i in out_text if not i.isdigit()])

        out_words = out_text.strip().lower().split()
        out_words = [word for word in out_words if len(word) > 3]
        out_words = [word for word in out_words if word not in self.__stopwords]

        return out_words

    
    def __compare_to_titles(self, string:str):

        string_to_compare = self.__preprocessing_text(string)

        comparison = dict()
        for title in self.__toc_titles:
            score = self.__tversky(title, string_to_compare)
            if score > 0.8 :
                comparison.update({score :   title})
        
        if comparison :
            max_score = max(comparison)
            return (self.__toc_titles_hierarchy[comparison[max_score]]['hierarchy'], self.__toc_titles_hierarchy[comparison[max_score]]['title'])
        else:
            return None

    def __parse_pdf(self):

        paragraphs = list()
        part_tite = ""
        sub_title = ""

        for page in self.__parsed_pdf:
            for p in page.get_text('blocks'):
                title = True
                p_text = p[4].split('\n')
                while self.__toc_titles and title and p_text :
                    title = self.__compare_to_titles(p_text[0])
                    if title :
                        hierarchy, title_string = title

                        if hierarchy == 1 :
                            part_tite = title_string
                        else :
                            sub_title = title_string
                        
                        text = text[1:]

                if p_text :

                    p_text = " ".join(p_text)

                    paragraph_length = re.subn(pattern=r"[^\w\s]|[\d]", repl='', string=p_text)[0]
                    paragraph_length = [word for word in nltk.word_tokenize(p_text) if word.lower() in self.__english_words]

                    if len(paragraph_length) >= 5 :
                        if paragraphs and paragraphs[-1][-1][-1] != "." :
                            paragraphs[-1][-1] = paragraphs[-1][-1] + " " + p_text.strip()
                        else :
                            paragraphs.append([part_tite, sub_title, p_text.strip()])
        
        if paragraphs :
            paragraphs_out = list()
            for paragraph in paragraphs :
                paragraph_out = dict()
                part_tite, sub_title, p_text = paragraph
                paragraph_out.update({
                    'content'   :   p_text,
                })

                if part_tite :
                    paragraph_out.update({
                    'part_title'   :   part_tite,
                })
                if sub_title :
                    paragraph_out.update({
                    'sub_title'   :   sub_title,
                })
                paragraphs_out.append(paragraph_out)

        self.paragraphs = paragraphs_out
```

```python
from elasticsearch import Elasticsearch
from typing import List
from dotenv import load_dotenv
load_dotenv('.env')
from pprint import pprint
from datetime import datetime
import os
from pathlib import Path
from sentence_transformers import SentenceTransformer

class TextEmbedding():

    model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
    
    def __init__(self) -> None:
        pass
    
    def embedded(self, text:str):
        self.model.encode(text)    

# class PDFFiles():
#     instances = 0

#     es = Elasticsearch(
#         hosts=[{'host': 'localhost', 'port': 9200, 'scheme': 'https'}],
#         ssl_assert_hostname='es01',
#         basic_auth=('elastic', os.getenv('ELASTIC_PASSWORD')),
#     # cert_reqs="CERT_REQUIRED",
#         ca_certs="./certifs/es/ca/ca.crt")

#     def __init__(self, pdf_id:str=None, pdf_issn:str=None, pdf_title:str=None) -> None:
#         self.instances += 1
#         if not (pdf_id or pdf_issn or pdf_title) :
#             e_text = "La class pdffile doit avoir au moin l'un des 3 arguements pour récupérer le PDF dans la base"
#             raise ValueError

#     def __del__(self):
#         if self.isinstance == 1:
#             self.es.close() 
#         self.instances += -1


class Search():
    instances = 0
    es = Elasticsearch(
        hosts=[{'host': 'localhost', 'port': 9200, 'scheme': 'https'}],
        ssl_assert_hostname='es01',
        basic_auth=('elastic', os.getenv('ELASTIC_PASSWORD')),
    # cert_reqs="CERT_REQUIRED",
        ca_certs="./certifs/es/ca/ca.crt")
    
    __text_embedding = TextEmbedding()

    __is_pdf_index = es.indices.exists(index="pdf_files")
    
    
    def __init__(self) -> None:
        self.instances += 1
        client_info = self.es.info()
        print("Connected to Elasticsearch !")
        pprint(client_info)
        self.pdf_index = ""
    
    def __create_pdf_index(self):
        self.es.indices.create(index="pdf_files")
        self.__is_pdf_index = True
        

    def create_index(self, index_name):
        self.es.indices.delete(index=index_name, ignore_unavailable=True)
        mappings={
            'properties': {
                'embedding': {
                    'type'          : 'dense_vector',
                    'dims'           : 384,
                    "index"         : True,
                    "similarity"    : "dot_product",
                }
            }
        }
        self.es.indices.create(index=index_name, mappings=mappings)
        self.index = index_name

    def __check_if_pdf_exists(self, document:Document):
        query_params = dict()

        pdf_document = document.get_pdf()

        fields = ["title", 'issn']

        for field in fields:
            if field in pdf_document:
                query_params.update({
                    field   :   pdf_document[field]
                })
        if query_params:
            query = {
                "query" :   {
                    "match" :   query_params
                }
            }
            sort_order = {"sort" : ["_score"]}
            result = self.es.search(index='pdf_files', query=query, sort=sort_order)
            results = result['hits']['hits']
            if results :
                document.es_pdf_id = results[0]["_id"]
                return True
            else :
                return False
        else :
            return None
        
    def __add_pdf(self, document:Document, created_on:datetime):
        pdf_doc = document.get_pdf()
        pdf_doc.update({
            "created_on"    :   created_on,
        })
        insert_result = self.es.index(index='pdf_files', document=pdf_doc)
        pdf_id = insert_result["_id"]
        if pdf_id :
            document.es_pdf_id = pdf_id
        else :
            e_text = "L'insertion du fichier PDF dans l'index pdf_files à échoué"
            raise ConnectionError(e_text)

    def add_documents(self, query_string:str, documents:List[Document], created_on:datetime, index:str=None):
        if not self.__is_pdf_index:
            self.__create_pdf_index()

        if not index:
            index = self.index
        
        if not isinstance(index, str):
            e_text = f"L'index doit être une string"
            raise ValueError(e_text)
        
        if not isinstance(documents, list):
            e_text = f"Le document doit être une liste de Document"
            raise ValueError(e_text)

        # if isinstance(paragraphs, dict):
        #     paragraphs = [paragraphs]
        # elif not isinstance(paragraphs, list):
        #     e_text = f"Les paragraphs à insérer doivent être des dictionnaires dans une liste"
        #     raise ValueError(e_text)
        

        operations = list()

        for document in documents:
            if document.doc_type == 'pdf':
                if not self.__check_if_pdf_exists(document):
                    self.__add_pdf(document=document, created_on=created_on)

            for paragraph in document():

                paragraph.update({
                    "created_on"    :   created_on
                })

                paragraph.update({
                    "user_query"    :   query_string,
                })

                doc_content = paragraph.get("content", None)
                if doc_content:
                    paragraph.update({
                        "embedding" :   self.__text_embedding.embedded(doc_content),
                    })

                operations.append({
                    'index' :   {'_index'   :   index}
                })

                operations.append(paragraph)
    
        return self.es.bulk(operations=operations, refresh='true')
    
    def __del__(self):
        if self.instances == 1:
            self.es.close()
        self.instances += -1
```

```python
from pymongo import MongoClient
import os
from dotenv import load_dotenv
load_dotenv("./.env")
load_dotenv("./dev.env", override=True)

def read_pwd_file(file_path:str) -> str:
    with open(file=file_path, mode="r") as pwd_file:
        pwd = pwd_file.readline().replace("\n", "")
    return pwd

mongo_client = MongoClient(
    host='localhost'
    ,port=27017
    ,username=read_pwd_file(os.getenv("MONGODB_USER_NAME_FILE"))
    ,password=read_pwd_file(os.getenv("MONGODB_USER_PWD_FILE"))
    ,tls=True
    ,tlsCAFile= "./certifs/mongodb/client/CA.pem"
    ,tlsCertificateKeyFile="./certifs/mongodb/client/client.pem"
)
```

```python
mongo_db = mongo_client.get_database(name=os.getenv("MONGO_DB_NAME"))
```
